```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for the attention layer
attention_data = {
    'Trial': list(range(1, 11)),
    'CPI Rate': [1.072, 1.022, 1.008, 0.989, 0.957, 0.916, 0.906, 0.964, 1.044, 0.914],
    'Front-End Bound': [34.6, 38.3, 40.3, 35.7, 34.0, 32.8, 30.2, 35.9, 36.1, 32.3],
    'Back-End Bound': [19.2, 29.6, 33.2, 24.9, 25.1, 28.5, 26.3, 27.2, 25.9, 29.2],
    'Memory Bound': [8.8, 12.9, 15.7, 11.3, 10.5, 11.4, 10.5, 11.1, 11.5, 12.9],
    'Core Bound': [10.4, 16.7, 17.5, 13.6, 14.5, 17.1, 15.8, 16.2, 14.5, 16.3],
    'Bad Speculation': [16.1, 2.7, 0.0, 11.4, 13.3, 12.1, 15.5, 3.9, 6.7, 9.2],
    'Retiring': [30.2, 29.3, 31.3, 28, 27.7,  26.7, 28, 32.9, 31.3, 29.3]
}

# Data for Feed-Forward Layer
feedforward_data = {
    'Trial': list(range(1, 11)),
    'CPI Rate': [0.929, 0.948, 1.020, 1.010, 0.987, 0.885, 0.895, 0.889, 0.933, 0.978],
    'Front-End Bound': [35, 34.1, 31.6, 37.7, 37.5, 35.3, 31.6, 31.5, 33.8, 38.6],
    'Back-End Bound': [34, 24.9, 21.6, 30.4, 31.9, 31.1, 30.2, 30, 25.4, 25.3],
    'Memory Bound': [13.8, 10.7, 10.4, 12.5, 12.7, 12.4, 13.2, 12.0, 11, 11.9],
    'Core Bound': [20.3, 14.3, 11.2, 17.9, 19.2, 18.7, 17, 18, 14.4, 13.4],
    'Bad Speculation': [4.3, 15.4, 17.6, 5.8, 2.9, 1.6, 12.2, 9.8, 14.1, 7.7],
    'Retiring': [2.6, 25.6, 29.2, 26, 27.8, 32, 26, 28.8, 26.7, 28.4]
}

# Create DataFrames
attention_df = pd.DataFrame(attention_data)
feedforward_df = pd.DataFrame(feedforward_data)

# Calculate average values excluding 'CPI Rate'
metrics_to_plot = ['Front-End Bound', 'Back-End Bound', 'Memory Bound', 'Core Bound', 'Bad Speculation', 'Retiring']
attention_avg = attention_df[metrics_to_plot].mean()
feedforward_avg = feedforward_df[metrics_to_plot].mean()

# Plot average values
plt.figure(figsize=(12, 6))

# Plot metrics
plt.plot(attention_avg.index, attention_avg.values, marker='o', linestyle='-', label='Attention Layer')
plt.plot(feedforward_avg.index, feedforward_avg.values, marker='o', linestyle='-', label='Feed Forward Layer')

# Add labels and legend
plt.title('Average Metrics Comparison (Excluding CPI Rate)')
plt.xlabel('Metric')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print conclusions for all metrics except CPI Rate
print("Conclusions for all metrics (excluding CPI Rate):")
for metric in metrics_to_plot:
    if attention_avg[metric] < feedforward_avg[metric]:
        print(f"- {metric}: Attention layer performs better.")
    elif attention_avg[metric] > feedforward_avg[metric]:
        print(f"- {metric}: Feed Forward layer performs better.")
    else:
        print(f"- {metric}: Both layers perform equally.")
```
